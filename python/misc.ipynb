```python
#convert PorCuEX2008.csv to gpml
import pandas as pd

import pygplates
from parameters import parameters
import Utils

data = pd.read_csv("../data/PorCuEX2008.csv")
display(data)

static_polygons = pygplates.FeatureCollection(parameters['static_polygons_file'])
rotation_model = pygplates.RotationModel(Utils.get_files(parameters['rotation_files']))
plate_ids = Utils.get_plate_id(data.LongitudeDecimal.tolist(), 
                               data.LatitudeDecimal.tolist(), 
                               static_polygons, 
                               rotation_model)
print(sorted(data.AgeMY))
data['PlateID'] = plate_ids
data.AgeMY.fillna(0,inplace=True)
point_features = []
for index,row in data.iterrows():
    point = pygplates.PointOnSphere(float(row.LatitudeDecimal),float(row.LongitudeDecimal))
    point_feature = pygplates.Feature()
    point_feature.set_geometry(point)
    point_feature.set_name(row.NameDeposit)
    point_feature.set_valid_time(float(row.AgeMY), pygplates.GeoTimeInstant.create_distant_future())
    point_feature.set_reconstruction_plate_id(int(row.PlateID))
    point_features.append(point_feature)
    
fc = pygplates.FeatureCollection(point_features)
fc.write('PorCuEX2008.gpmlz')
print('done!')
```

```python
# Given two points, get third point of triangle.
import Quaternions
import math

PA_lon_lat = (-12.194557, 13.423369)
PB_lon_lat = (11.161865, -11.856508)

PA_xyz = (Quaternions.lat_lon_to_cart(math.radians(PA_lon_lat[1]), math.radians(PA_lon_lat[0])))
PB_xyz = (Quaternions.lat_lon_to_cart(math.radians(PB_lon_lat[1]), math.radians(PB_lon_lat[0])))

AB_middle_xyz = ((PA_xyz[0] + PB_xyz[0])/2, (PA_xyz[1] + PB_xyz[1])/2, (PA_xyz[2] + PB_xyz[2])/2)
AB_middle_xyz = Quaternions.normalize(AB_middle_xyz)

print(AB_middle_xyz)

AB = (PA_xyz[0] - PB_xyz[0], PA_xyz[1] - PB_xyz[1], PA_xyz[2] - PB_xyz[2])
AB =  Quaternions.normalize(AB)

print(AB)

middle_lat, middle_lon = Quaternions.cart_to_lat_lon(AB_middle_xyz[0], AB_middle_xyz[1], AB_middle_xyz[2])
pole_lat, pole_lon = Quaternions.cart_to_lat_lon(AB[0],AB[1],AB[2])

middle_lat = math.degrees(middle_lat)
middle_lon = math.degrees(middle_lon)
pole_lat = math.degrees(pole_lat)
pole_lon = math.degrees(pole_lon)

with open('ring.gmt','w+') as f:
    f.write(f"{middle_lon:.2f} {middle_lat:.2f}\n")
    f.write(f"{pole_lon:.2f} {pole_lat:.2f}\n")
    f.write(f"{PA_lon_lat[0]:.2f} {PA_lon_lat[1]:.2f}\n")
    f.write(f"{PB_lon_lat[0]:.2f} {PB_lon_lat[1]:.2f}\n")
    for angle in range(360):
        lat, lon = Quaternions.rotate([middle_lat, middle_lon],
                                      [pole_lat, pole_lon],
                                      angle)
        f.write(f"{lon:.2f} {lat:.2f}\n")
        
print('done')
```
