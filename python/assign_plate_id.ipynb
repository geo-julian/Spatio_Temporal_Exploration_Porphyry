```python
import pandas as pd
import pygplates
import Utils

input_file_name = 'deposit_positives.csv'
config_file_name = 'config.AREPS.json'


Utils.load_config(config_file_name)
Utils.print_parameters()

data = pd.read_csv(input_file_name)
#display(data)

lons=data['lon']
lats=data['lat']
static_polygons = pygplates.FeatureCollection(Utils.get_parameter('static_polygons_file'))
rotation_model = pygplates.RotationModel(Utils.get_files(Utils.get_parameter('rotation_files')))
plate_ids = Utils.get_plate_id(lons.tolist(), 
                               lats.tolist(), 
                               static_polygons, 
                               rotation_model)

print(plate_ids)
data['plate_id'] = plate_ids
data.to_csv(input_file_name, index=False)
display(data)
print(f'New data has been saved in {input_file_name}.')
```
