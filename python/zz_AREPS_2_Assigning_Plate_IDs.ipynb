```python
import pygplates
import pandas as pd
import numpy as np
from parameters import parameters
import Utils as Utils
import math
import csv


NA_positives = pd.read_csv('input_data/NA_positives.csv')
NA_negatives = pd.read_csv('input_data/NA_negatives.csv')
SA_positives = pd.read_csv('input_data/SA_positives.csv')
SA_negatives = pd.read_csv('input_data/SA_negatives.csv')


point_features_NA_positives = []
for index,row in NA_positives.iterrows():
    point = pygplates.PointOnSphere(float(row[1]),float(row[2]))
    point_feature = pygplates.Feature()
    point_feature.set_geometry(point)
    point_feature.set_geometry_import_time(int(row[3]))
    point_features_NA_positives.append(point_feature)

point_features_NA_negatives = []
for index,row in NA_negatives.iterrows():
    point = pygplates.PointOnSphere(float(row[1]),float(row[2]))
    point_feature = pygplates.Feature()
    point_feature.set_geometry(point)
    point_feature.set_geometry_import_time(int(row[3]))
    point_features_NA_negatives.append(point_feature)
    
point_features_SA_positives = []
for index,row in SA_positives.iterrows():
    point = pygplates.PointOnSphere(float(row[1]),float(row[2]))
    point_feature = pygplates.Feature()
    point_feature.set_geometry(point)
    point_feature.set_geometry_import_time(int(row[3]))
    point_features_SA_positives.append(point_feature)

point_features_SA_negatives = []
for index,row in SA_negatives.iterrows():
    point = pygplates.PointOnSphere(float(row[1]),float(row[2]))
    point_feature = pygplates.Feature()
    point_feature.set_geometry(point)
    point_feature.set_geometry_import_time(int(row[3]))
    point_features_SA_negatives.append(point_feature)
    
static_polygons = pygplates.FeatureCollection(parameters['static_polygons_file'])
rotation_model = pygplates.RotationModel(Utils.get_files(parameters['rotation_files']))

partitioned_NA_positives = pygplates.partition_into_plates(static_polygons, rotation_model, point_features_NA_positives)
partitioned_NA_negatives = pygplates.partition_into_plates(static_polygons, rotation_model, point_features_NA_negatives)
partitioned_SA_positives = pygplates.partition_into_plates(static_polygons, rotation_model, point_features_SA_positives)
partitioned_SA_negatives = pygplates.partition_into_plates(static_polygons, rotation_model, point_features_SA_negatives)

features_NA_positives = pygplates.FeatureCollection(partitioned_NA_positives)
features_NA_negatives = pygplates.FeatureCollection(partitioned_NA_negatives)
features_SA_positives = pygplates.FeatureCollection(partitioned_SA_positives)
features_SA_negatives = pygplates.FeatureCollection(partitioned_SA_negatives)
```
